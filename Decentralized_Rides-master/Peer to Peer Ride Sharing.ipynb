
## The Problem with Uber

![alt text](http://www.taxifarefinder.com/newsroom/wp-content/uploads/2016/06/uber-by-the-numbers-RideGuru-infographic.jpg "Logo Title Text 1")

- The 66 billion valued Uber hopes to become the world’s dominant transportation and logistics network.
- Uber drivers have been complaining that the gap between the fare a rider pays and what the driver receives is getting wider
- Uber's official statement stated that they're charging some passengers more because they need the extra cash
- The company applies machine-learning techniques to estimate how much groups of customers are willing to shell out for a ride.
- For instance, someone traveling from a wealthy neighborhood to another tony spot might be asked to pay more than another person heading to a poorer part of town, even if demand, traffic and distance are the same.
- "Be Your own Boss"? No. Real entrepreneurs are free to compete in an open marketplace where they can set their own pricing and the terms of the transaction. 
- Drivers face massive pay cuts overnight with no warning, and must conduct business ‘their way’ or not at all
- Centralization makes Uber vulnerable to regulatory actions, and there is the possibility that it could be shut down by the government at any time (and has been in many cities e.g Austin)

## Hmm. BLOCKCHAIN?!?!? 

![alt text](https://bitcab.io/images/DldJlWgp7zQ.jpg "Logo Title Text 1")

- Building applications on the blockchain gives them 3 (sometimes) useful properties: 

1. Immutability - nothing written to the blockchain can be changed 
2. Persistence - everyone has a copy so you can't lose all your data 
3. Trustlessness - no need to rely on a central authority for information

- There are a lot of edge cases in an application like Uber. If you get charged for a ride but never picked up, you wouldn't want a transaction like that to be permanent.
- Almost any application CAN be built on the blockchain, but a lot of thought should be given to whether or not it SHOULD
- A blockchain could provide us with some incredible properties to help aid in the formation of a new business model for ride sharing.

TL;DR it could free drivers to:

![alt text](https://cdn-images-1.medium.com/max/1600/0*9p__iOTiZbimQaLa. "Logo Title Text 1")

- Earn equity
- Set their own rates
- Negotiate and transact directly with riders
- Build direct relationships with riders for a stable and recurring customer base
- Self-organize into ‘pods’, co-ops, or their own businesses

## SnagRide

![alt text](https://42xtjqm0qj0382ac91ye9exr-wpengine.netdna-ssl.com/wp-content/uploads/2018/01/snagride-1024x241.jpg "Logo Title Text 1")

![alt text](https://www.autonomousvehicletech.com/ext/resources/AVT/2018/01-January/20180103-Snag-Ride-image_web.jpg?1515015820 "Logo Title Text 1")

- Smart contracts are used for
1. Drivers posting their destination and departure details. 
2. Drivers will state when and where they are going. 
3. Users will be able to sign up for a ride by paying their share of the estimated cost and then wait to be picked up by the driver
- Smart contracts contain all the information such  as rider name, age, location, ride price and where he would like to travel. 
- All data is  encrypted on the Ethereum blockchain. 
- A driver offering a ride will view that information, schedule a ride and then choose a date and time to confirm. The driver will then confirm the
ride. 
- Once the driver confirms the ride a new transaction is initiated that will broadcast the confirmation and driver information to the Ethereum blockchain.
- All rides executed through this proprietary intelligent smart contract
- This smart contract will contain information on both the rider and driver. This information will be used to match the rider with their most prefered modes of transportation.
- The token is called MILE. Its used to pay drivers and allows their AI to match riders and drivers through analyzing user preferences data, including price limits, general routing preferences, time of rides, type of vehicles etc.

## Arcade City

![alt text](https://thisis.swarm.city/images/img_section_how.png "Logo Title Text 1")

- Began in late December 2015 in defiance of transportation regulations in Portsmouth, New Hampshire which had made Uber illegal.
- To compensate for the imminent transportation shortage, a small group of current and former Uber drivers organized under the name Arcade City to provide peer-to-peer rides to partygoers free of charge.
- The basic ‘v0’ Arcade City mobile app was launched in mid-February and was live for two months, with rides given in 28 states and Australia. 
- Drivers could set their own rates, advertising their personal policies and payment methods on a customizable driver profile. 
- Its an Ethereum-based app for web, Android and iOS.
